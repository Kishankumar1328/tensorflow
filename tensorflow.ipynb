```python
import tensorflow as tf
import numpy as np

a=tf.constant(13)
b=tf.constant(29)

sum=a+b

print("The sum of{} and {} is:{}".format(a.numpy(),b.numpy(),sum.numpy()))
```

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(0)
x=2*np.random.randn(100,1)
y=4 + 3 *x+np.random.randn(100, 1 )

model=tf.keras.Sequential([
    tf.keras.layers.Input(shape=(1,)),
    tf.keras.layers.Dense(units=13,activation="linear")
])

model.compile(optimizer='sgd',loss='mean_squared_error')

model.fit(x,y,epochs=1329,verbose="auto")
```

```python
prediction=model.predict(x)
```

```python
plt.scatter(x,y,label="Original data")
plt.plot(x,prediction,"-r",label="Linear Regression")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
```

```python
np.random.seed(0)
x=2*np.random.rand(100,1)
y=4+3*x+np.random.randn(100,1)

model=tf.keras.Sequential([
    tf.keras.layers.Input(shape=(1,)),
    tf.keras.layers.Dense(units=1,activation='linear')])

model.compile(optimizer='sgd',loss="mean_squared_error")

model.summary()

history=model.fit(x,y,epochs=160,verbose="auto")
```

```python
plt.plot(history.history['loss'])
plt.xlabel('Epoch')
plt.ylabel("Mean squared Error Loss")
plt.title(' Training Loss Over Epochs')
```
